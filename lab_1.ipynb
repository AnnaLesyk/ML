```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 12, 1000)

plt.plot(x, x**2, color = 'blue', label = 'x^2')
plt.plot(x, 10 * np.sin(x), color = 'red', label = '10*sin(x)')
plt.plot(x, 10 * np.sin(x) + x**2, color = 'green', label = '10*sin(x)+x^2')

plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

def f(x, y):
    return (x**2 + y**2 )

x = np.linspace(-1, 1, 30)
y = np.linspace(-1, 1, 30)
#z = np.linspace(-1, 1, 30)

X, Y = np.meshgrid(x, y)
r = f(X, Y)
R = r**2

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.contour3D(X, Y, R, 50, cmap='binary')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#ax.plot_surface(X, Y, Z, color = 'black', rstride = 1, cstride = 1)
plt.show()
```
